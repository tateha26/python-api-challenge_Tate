```python
import tweepy
import numpy as np
import json
import random
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
import requests
import time
import sys
import matplotlib.lines as mlines

from citipy import citipy
from config_Tate import consumer_key, consumer_secret, access_token, access_token_secret
from config_Tate import weather_api
```

```python
# Setup Tweepy API Authentication
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, parser=tweepy.parsers.JSONParser())
```

```python
# Import and Initialize Sentiment Analyzer
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
target_terms = "@lighteves","@theBadChai", "@IAmDartanion", "@justinthomasva"

compound_list = []
positive_list = []
negative_list = []
neutral_list = []
sentiments = []
```

```python
for user in target_terms:
    counter=0
    #tweets = api.search(user, count=500, result_type="recent")
    tweets= api.user_timeline(user, count=100)
    
    for tweet in tweets:   
        compound = analyzer.polarity_scores(tweet["text"])["compound"]
        pos = analyzer.polarity_scores(tweet["text"])["pos"]
        neu = analyzer.polarity_scores(tweet["text"])["neu"]
        neg = analyzer.polarity_scores(tweet["text"])["neg"]
        tweets_ago=counter
        #date=time.strftime("%d/%m/%y")
        
        sentiments.append({"User": user, "Compound": compound, 
                           "Positive": pos, "Negative": neg, 
                           "Neutral": neu, "Tweets Ago": counter})
        counter=counter+1
twitter_df = pd.DataFrame(sentiments)
twitter_df.head()
```

```python
for user in target_terms:
    twitter_df_2=twitter_df.loc[twitter_df["User"]==user]
    plt.scatter(twitter_df_2["Tweets Ago"], twitter_df_2["Compound"])
    plt.title("Sentiment Analysis of Tweets: All Users")
    plt.xlabel("Tweets Ago")
    plt.ylabel("Tweets Polarity")
    lighteves = mlines.Line2D([], [], color='blue',
                                 markersize=10, label='@lighteves')
    theBadChai = mlines.Line2D([], [], color='red',
                                 markersize=10, label='@theBadChai')
    IAmDartanion = mlines.Line2D([], [], color='pink',
                                 markersize=10, label='@IAmDartanion')
    justinthomasva = mlines.Line2D([], [], color='orange',
                                 markersize=10, label='@justinthomasva')
    plt.legend(handles=[lighteves, 
                        theBadChai, 
                        IAmDartanion, 
                        justinthomasva]
              )
plt.show()
```
